```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../input/musk-dataset/musk_csv.csv')
data.head()
```

#since id and confronation names are unique ,we can remove them.


```python
data = data.drop(['molecule_name','ID','conformation_name'],axis=1)
```

```python
data.head()
```

```python
sns.countplot(data['class'])
plt.show()
```

```python
data.isnull().values.any()
```

There are no null values present in the dataset

### Balancing the dataset

```python
data['class'].value_counts()
```

```python
data_0 = data[data['class'] == 0].iloc[:len(data[data['class'] == 1]),:]
data_1 = data[data['class'] == 1]
```

**Concatenating data_0 and data_1 to form balanced dataset**

```python
new_data = pd.concat([data_0,data_1])
new_data = new_data.sample(frac = 1)
```

### Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(new_data, new_data['class'], test_size = 0.20, random_state = 101)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### Using Knn

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train, Y_train)
```

#### Making prediction

```python
y_pred = classifier.predict(X_test)
```

#### Performance metrics on test set

```python
from sklearn.metrics import classification_report
cls = classification_report(Y_test, y_pred)
print(cls)
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(Y_test, y_pred)
print(cf)
```
